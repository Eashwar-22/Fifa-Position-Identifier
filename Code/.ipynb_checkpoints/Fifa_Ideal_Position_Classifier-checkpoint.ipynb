```python
df_orig = pd.read_csv('../Data/players.csv')
```

**Drop unnecessary columns - that are not needed for predicting recommended positions**

```python
df_orig.columns=[i.replace('int_','').replace('str_','').replace('dt_','') for i in df_orig.columns]
columns_to_drop = ['player_id','player_name','date_of_birth','overall_rating','potential_rating',\
                   'value','wage','team_id','nationality','preferred_foot',\
                   'international_reputations','body_type','player_speciality','trait',\
                   'positions','best_overall_rating']

data = df_orig.drop(columns=columns_to_drop)
data.head()
```

**Dealing with Missing Data**

```python
print(f"Number of rows with missing values : {len(data)-len(data.dropna())} out of {len(data)} rows")

plt.figure(figsize=(15,5))
sns.heatmap(data.isnull());
```

### Pre-processing to be done on columns
- 'work-rate' : remove spaces in string<br>
- 'positions' : (Target variable)          
     - remove spaces <br>
     - convert the comma seperated string into list <br>
     - split them into columns <br>


```python
data['work_rate']=data['work_rate'].apply(lambda x : x.replace(' ',''))
data['positions']=data['positions'].apply(lambda x : x.replace(' ','').split(','))
```

```python
X = data.drop(['best_position','positions'],axis=1)

target=pd.DataFrame(data['positions'].copy())
all_positions = list(data['best_position'].unique())

for i in all_positions:
    target[i]=target['positions'].apply(lambda x: int(i in x))
    
y = target.drop(['positions'],axis=1)

print(X.shape,y.shape)
```

```python
def plot_dist(data,col):
    """
    Plot distributions of a continuous variable
    """
    fig,ax1=plt.subplots(1,1,figsize=(15,7))
    ax1.set_title(f'Distribution of {col} with Skewness = {round(data[col].skew(),2)}',fontsize=16)
    sns.distplot(data[col],ax=ax1);
#     sns.boxplot(data=data,x=col,ax=ax2);
```

**Encoding & Feature Scaling**

```python
def transform_(X):
    # One hot encoding for Work Rate
    wr = pd.get_dummies(X['work_rate'],drop_first=True) 

    # MinMax Scaling the numerical attributes
    X_scaled = X[[j for j in X.columns if j!='work_rate']].copy()
    scaler = MinMaxScaler()
    save_cols = X_scaled.columns
    X_scaled=np.array(X_scaled)
    X_scaled= scaler.fit_transform(X_scaled)
    X_new = pd.DataFrame(X_scaled,columns=save_cols)
    X_new = pd.concat([X_new,wr],axis=1)
    return X_new
    
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_new,y,test_size=0.2)
for j in [X_train,X_test,y_train,y_test]:
    print(j.shape)
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)

# predict
predictions = classifier.predict(X_test)

accuracy_score(y_test,predictions)
```
